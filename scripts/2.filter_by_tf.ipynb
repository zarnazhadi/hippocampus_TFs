```python
# Imports sample expression data from files, selects one transription factor and merges with database of tissue types
import os
import pandas as pd
import numpy as np
```

```python
# import expression data

# fields = ['Name','Description', 'GTEX-1128S-2726-SM-5H12C','GTEX-1192X-0011-R5a-SM-DNZZA','GTEX-1128S-2826-SM-5N9DI']
exp = pd.read_csv("/Users/zarnazhadi/Desktop/hippocampus_TFs/preliminary_data/RNA_seq.txt", sep= '\t', header= 2)

# filter by row for preferred tf
tf = 'DDX11L1'
is_tf =  exp['Description']== tf
exp_tf = exp[is_tf]

# formatting

# remove first two columns 
del exp_tf['Name']
del exp_tf['Description']

# transpose 
exp_tf = exp_tf.transpose()
col_name = tf + '_expression'
exp_tf.columns = [col_name]

# convert row names to column sample in exp_tf
exp_tf.index.name = 'sample'
exp_tf.reset_index(inplace=True)
```

```python
# import tissue_type txt file
with open("/Users/zarnazhadi/Desktop/hippocampus_TFs/preliminary_data/tissue_types.txt") as f:
    content = f.readlines()
    
# formatting
content = [x.strip() for x in content] # remove \n
tissue_types =[s.replace('\t', ' ') for s in content] # remove \t

# filter by preferred tissue e.g. Brain
tissue = 'Brain'
tissue_list = [s for s in tissue_types if tissue in s]
```

```python
# split into sample name and tissue type
start = " - "
end = "0"
bt = []
for brain in tissue_list:
    b = brain.split()[0] # GTEX sample
    t = (brain.split(start))[1].split(end)[0] # tissue type
    bt.append(b)
    bt.append(t)

bt # list of brain tissue
```

```python
# make copy of exp_tf and filter for only brain tissue - for reference if needed
# exp_brain = exp_tf.copy(deep = True)
# exp_brain
bt
```

```python
# convert bt to a database
samples = []
tissue = []
for index, b in enumerate(bt):
    if index == 0 or index %2 == 0:
        print(index, b)
        samples.append(b)
    elif index %2 != 0:
        print(index, b)
        tissue.append(b)
```

```python
brain_tissue = pd.DataFrame(
    {'sample': samples,
     'tissue_type': tissue,
    })
```

```python
# merge brain_tissue and exp_tfs
merged = pd.merge(left= exp_tf, right=brain_tissue, left_on='sample', right_on='sample')
merged.to_csv (r"/Users/zarnazhadi/Desktop/hippocampus_TFs/expression_data/DDX11L1_expression_data.csv", index = False, header=True)
```
