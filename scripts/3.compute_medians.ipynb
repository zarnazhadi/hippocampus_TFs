```python
import os
import pandas as pd
import numpy as np

dat = pd.read_csv("/Users/zarnazhadi/Desktop/hippocampus_TFs/expression_data/10_TFs_expression_data.csv", sep= ',', header=0) # change working directory
```

```python
del dat['sample'] # delete sample column
```

```python
medians = dat.groupby(['tissue_type']).median() # median
means= dat.groupby(['tissue_type']).mean() # mean
```

```python
means
```

```python
medians = medians.transpose()
```

```python
# make new comparison table by subsetting medians with only hippocampus
comparison = pd.DataFrame(medians['Hippocampus '])

# compute difference for each column from hippocampus- with true/false and hippocampus-median of other tissue
for column in medians.columns:
    if column != 'Hippocampus ':
        #print('true')
        colname1 = column + " > Hippocampus"
        print(colname1)
        # find out if median is larger than hippocampus median
        comparison[colname1]  =medians.apply(lambda  x: True if x[column]>x['Hippocampus '] else False ,axis=1)
        # compute difference between the two
        colname2 = "Hippocampus - " + column 
        print(colname2)
        comparison[colname2]  =medians.apply(lambda  x: x['Hippocampus ']-x[column],axis=1)
# comparison data frame should now include boolean and difference values compared to hippocampus
```
