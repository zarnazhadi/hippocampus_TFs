```python
medians = dat.groupby(['tissue_type']).median() # median
means= dat.groupby(['tissue_type']).mean() # mean
```

```python
medians = medians.transpose() # formatting

#remove rows with all zeros
#medians = medians[(medians.T != 0).any()]
```

```python
def ratio(x):
    try:
        float(x[column]) / float(x['Hippocampus ']))
    except ZeroDivisionError:
        0
```

```python
# make new comparison table by subsetting medians with only hippocampus

comparison = pd.DataFrame(medians['Hippocampus '])
only_true = pd.DataFrame(medians['Hippocampus '])

# compute difference for each column from hippocampus- with true/false and hippocampus-median of other tissue

for column in medians.columns:
    if column != 'Hippocampus ':
        
        # find out if median is larger than hippocampus median
        colname1 = column + " > Hippocampus"
        comparison[colname1]  =medians.apply(lambda  x: True if x[column] > x['Hippocampus '] else False ,axis=1)
        
        # compute difference between the two
        colname2 = column + " - Hippocampus"
        comparison[colname2]  =medians.apply(lambda  x: x[column] - x['Hippocampus '],axis=1)
        # print(comparison.sort_values(colname2))
        
        # ratio
        colname3 = column + "/Hippocampus"
        comparison[colname3] =medians.apply(ratio,axis=1)  
```

```python
comparison
```

```python
# comparison_copy = comparison.copy(deep = True)
value = comparison.values
tf = comparison.index
tissue = comparison.columns

#  Get indices where such threshold is crossed; avoid diagonal elems
R,C = np.where(np.triu(value,1)>0)

# Arrange those in columns and put out as a dataframe
out_value = np.column_stack((tf[R],tissue[C],value[R,C]))
only_true = pd.DataFrame(out_value,columns=[['tf','col_name','value']])
only_true.columns = ['tf', 'tissue-hippocampus', 'value']
```

```python
only_true['rank'] = only_true['value'].rank(ascending=False).astype(int)
```
